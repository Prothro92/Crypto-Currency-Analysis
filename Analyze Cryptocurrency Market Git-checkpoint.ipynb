# Bitcoin Crytpocurrency Analysis

## Import Dataframes

```python
import json
import time
from urllib.request import urlopen
import pandas as pd
import datetime as dt
```

```python
def Bitcoin_DF():
    b=[]
    for i in currencies:
        url = 'https://poloniex.com/public?command=returnChartData&currencyPair=BTC_'+str(i)+'&start=1405699200&end=9999999999&period=86400'
        #url from Poloniex which returns data for bitcoin currency pair from a start and end point. Start represents unix data where data started being recorded
        #Period 86400 represents 24 hour period
        json_strings = urlopen(url).read()
        the_data = json.loads(json_strings)
        i = pd.DataFrame()
        n = 0
        while n <= len(the_data)-1:
            for k in the_data[n]:
                i[str(k).title()]=[item[str(k)] for item in the_data]
                n = n+1 #while loop which takes dictonary keys and values returned from the url and adds them into a dataframe
        i['Date']=i['Date'].apply(lambda x: dt.datetime.fromtimestamp(int(x)).strftime('%Y-%m-%d'))
        #Changes unixtime stamp into regular datetime format
        i['Daily Difference']=i['Close']-i['Open']
        i.set_index('Date',inplace=True)
        i = i.rename(columns={'Quotevolume':'Quote_Volume','Weightedaverage':'Weighted_Average'})     
        b.append(i)
    return b
currencies = ['ETH','LTC','XMR']
Cryptos=Bitcoin_DF()
Crypto_Stocks = pd.concat([Cryptos[0],Cryptos[1],Cryptos[2]],join='outer',keys=currencies,axis=1)
#concats all stocks into one single dataframe
Crypto_Stocks.columns.names=['Crypto Name','Stock Info']
```

```python
Crypto_Stocks.head()
```

## Calculating Daily Returns of Bitcoin Currency Pairs With The Above Dataframe

```python
def Bitcoin_Returns(Crypto):
    BTC_Returns = pd.DataFrame()
    for i in currencies:
        BTC_Returns[str(i)+' Returns']=Crypto[i]['Close'].pct_change(1)
    return BTC_Returns
BTC_Returns = Bitcoin_Returns(Crypto_Stocks)
```
